Compare the Average Vapor Pressure of the two places.     

Vapor Pressure definition: It is the amount of pressure in millibars exerted by the water vapor in the air.    

It is natural to make the assumption that the Average Vapor Pressure of Pacific Grove is higher than that of Coalinga since Pacific Grove is a coastal city while Coalinga is located in California's central valley.

```python
#  Average Vapor Pressure of Pacific Grove 
caw_vapor = caw_df.iloc[:,[8]]
caw_vapor
```

```python
print(type(caw_vapor.index))
```

```python
# Monthly and quarterly sum of Average Vapor Pressure, Pacific Grove
print("Monthly Sum of Average Vapor Pressure, Pacific Grove")
print()
caw_vapor_monthly_sum = caw_vapor.resample('M').sum()
print(caw_vapor_monthly_sum)
print('-----------------------------------------------------')
print("Quarterly Sum of Average Vapor Pressure, Pacific Grove")
print()
print(caw_vapor.resample('Q').sum()/3)
```

```python
# plt.title('Monthly Sum of Average Vapor Pressure, Pacific Grove')
# plt.grid(True)
# caw_plot = plt.plot(caw_vapor_monthly_sum.index, caw_vapor_monthly_sum['Average Vapor Pressure   (mBars)   (kPa)'],linewidth=3)
```

```python
#  Average Vapor Pressure of Coalinga
caw_vapor_compare = caw_df_compare.iloc[:,[8]]
caw_vapor_compare
```

```python
print(type(caw_vapor_compare.index))
```

```python
# Monthly and quarterly sum of Average Vapor Pressure, Coalinga
print("  Monthly Sum of Average Vapor Pressure, Coalinga")
caw_vapor_compare_monthly_sum = caw_vapor_compare.resample('M').sum()
print(caw_vapor_compare_monthly_sum)
print('-----------------------------------------------------')
print("  Quarterly Sum of Average Vapor Pressure, Coalinga")
print(caw_vapor_compare.resample('Q').sum()/3)
```

```python
plt.plot(caw_vapor_monthly_sum.index, caw_vapor_monthly_sum['Average Vapor Pressure   (mBars)   (kPa)'],label='Pacific Grove',linewidth=3)
plt.plot(caw_vapor_compare_monthly_sum.index, caw_vapor_compare_monthly_sum['Average Vapor Pressure   (mBars)   (kPa)'],label='Coalinga',linewidth=3)
plt.title('Pacific Grove vs Coalinga: Monthly Sum of Average Vapor Pressure')
plt.ylabel('Average Vapor Pressure')
plt.xlabel('Month')
plt.grid()
plt.legend()
```

From the graph we can prove that our assumption is right.    
Each month, the sum of Avarage Vapor Pressure of Pacific Grove is about 300 kPa higher than that of Coalinga.     
We also noticed that the Avarage Vapor Pressure of Pacific Grove and Coalinga both reaches the peak during the summer.
